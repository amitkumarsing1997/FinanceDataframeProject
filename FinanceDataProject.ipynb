```python
##now I create list of the ticker symbols in string format as well in alphabetical order
tickers = ['BAC','C','GS','JPM','MS','WFC']
tickers
```

```python
#now I used pd.concat the bank dataframes together to a single data
#frame called bank_stocks .now I set the keys argument according to tickers list.Here I also take care axis.
```

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
bank_stocks
```

```python
# get top data for understanding
bank_stocks.head()
```

```python
#amit ,for better understanding and nice view of data creates columns names
bank_stocks.columns.names=['Bank Ticker Name','Stock Info']
```

```python
bank_stocks.head()
```

```python
##now I will find out what is the max close price for each bank's througout the time period

for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
#another way to find out using multi-level indexing and Using .xs() function
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()

```

```python
##now I will find out what is the min close price for each bank's througout the time period

for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].min())
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
#amit_code
for tick in tickers:
    returns[tick + ' Return '] = bank_stocks[tick]['Close'].pct_change() 
returns.head()
```
